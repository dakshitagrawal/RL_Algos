```python
import random
import gym
import collections
import numpy as np
from matplotlib import pyplot as plt
```

```python
def select_greedy_action(table, state):
    
    value, action = best_action_value(table, state)
    
    return action

def select_eps_greedy_action(table, state, n_actions):
    
    value, action = best_action_value(table, state)
    
    prob = random.random()
    
    if prob < epsilon:
        return random.randint(0, n_actions-1)
    else:
        return action
    
def best_action_value(table, state):
    best_action = 0
    best_value = 0
    
    for i in range(n_actions):
        if table[(state, i)] > best_value:
            best_action = i
            best_value = table[(state, i)]
            
    return best_value, best_action
```

```python
def new_q_value(table, reward, state, action, state1):
    best_value, _ = best_action_value(table, state1)
    
    q_target = reward + GAMMA * best_value
    
    q_error = q_target - table[(state, action)]
    
    table[(state, action)] += ALPHA * q_error
```

```python
def test_game(env, table):
    reward_games = []
    
    for _ in range(TEST_EPISODES):
        state = env.reset()
        rewards = 0
        
        while True:
            next_state, reward, done, _ = env.step(select_greedy_action(table, state))
            state = next_state
            rewards += reward
            
            if done:
                reward_games.append(rewards)
                break
    return np.mean(reward_games)      
```

```python
def final_play(env, table):
    state = env.reset()
    env.render()
    
    total_reward = 0
    done = False
    
    while done!=True:
        state, reward, done, _ = env.step(select_greedy_action(table, state))
        total_reward += reward
        env.render()
        
    print(total_reward)
```

```python
GAMMA = 0.95
ALPHA = 0.8
TEST_EPISODES = 100
MAX_GAMES = 100000
epsilon = 1.0
EPSILON_DECAY = 0.9993
```

```python
env = gym.make('FrozenLake-v0')
state = env.reset()

n_states = env.observation_space.n
n_actions = env.action_space.n

games_count = 0
rewards = 0
test_rewards_list = []

table = collections.defaultdict(float)

while games_count < MAX_GAMES:
    
    action = select_eps_greedy_action(table, state, n_actions)

    next_state, reward, done, _ = env.step(action)

    new_q_value(table, reward, state, action, next_state)

    state = next_state
    rewards += reward
    
#     print(rewards)
    
    if done:
#         print(rewards)
        if games_count % 1000 == 0:
            test_reward = test_game(env, table)
            print('\tEp:', games_count, 'Test reward:', test_reward, np.round(epsilon,2))
            test_rewards_list.append(test_reward)
            
        epsilon = epsilon * EPSILON_DECAY
        state = env.reset()
        rewards = 0
        games_count += 1

plt.figure(figsize = (20, 10))
plt.xlabel("games_count")
plt.ylabel("rewards")
plt.plot(test_rewards_list)
plt.show()
```

```python
final_play(env, table)
```
